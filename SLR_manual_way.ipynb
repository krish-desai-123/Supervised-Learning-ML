```python
import pandas as pd
import matplotlib.pyplot as plt
```

# <i><b>SLR - Manual Way

```python
 x = pd.Series([1, 2, 3])
 y = pd.Series([2, 4, 5])
```

```python
x_mean = x.mean()
y_mean = y.mean()

numerator  = sum((x - x_mean) * (y - y_mean))
denominator = sum((x - x_mean) ** 2)

beta_1 = numerator / denominator
beta_0 = y_mean - (beta_1 * x_mean)

print(f'Mean of x: {x_mean:.2f} | Mean of y: {y_mean:.2f}')
print(f'Numerator: {numerator:.2f} | Denominator: {denominator:.2f}')
print(f'Beta_1 (Slope): {beta_1:.2f}\nBeta_0 (Intercept): {beta_0:.2f}')

print(f'Final Equation: y = {beta_0:.2f} + {beta_1:.2f}x')
```

```python
y_pred = beta_0 + beta_1 * x

residuals = y - y_pred

sse = sum(residuals ** 2)

print(f'Predicted y values: \n{y_pred}')
print(f'Residuals: \n{residuals}')
print(f'SSE: {sse:.2f}')
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.show()
```
